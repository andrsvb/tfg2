PRUEBAS CAMBIAR \QR POR CODIGO QUE INSERTE QRs

```python
from texsurgery.texsurgery import TexSurgery
```

```python
def QR(# self,
       exam, variant,
       scale='0.8',
       angle='0',
       color='black',
       posx='0.85', posy='0.95',
       anchor='rb',
       qr_size = '0.08'):
    text_config = '\\usepackage{qrcode}\n'
    text_config = text_config + '\\usepackage[' + \
        'scale=' + scale + \
        ',angle=' + angle + \
        ',color=' + color + \
        ',pos={' + posx + '\\paperwidth,' + posy + '\\paperheight}' + \
        ',anchor=' + anchor + ']{draftwatermark}'
    text_config = text_config + '\n' \
        + '\\SetWatermarkText{\qrcode[height=' + qr_size + '\\paperwidth]{'+ exam + ','+ variant+ ',\\thepage}}'



    print(text_config)
    return text_config
```

PRUEBAS LEER QRs

```python
file = 'solution_11111.pdf'
from pdf2image import convert_from_path
import cv2
import os
```

```python
pages = convert_from_path(file, 250)
tmp_img = 'outimg_tmp.png'
qr_detect = cv2.QRCodeDetector()
data = {'values': [], 'points': [], 'qrcodes': []}
for page in pages:
    page.save(tmp_img)
    img = cv2.imread(tmp_img)
    value, points, straight_qrcode = qr_detect.detectAndDecode(img)
    data['values'].append(value)
    data['points'].append(points)
    data['qrcodes'].append(straight_qrcode)
os.remove(tmp_img)
data
```
